```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd
```

```python
# Create a path variable to the data.
school_data_to_load = Path("../Resources/schools_complete.csv")
student_data_to_load = Path("../Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
school_data.head()
```

```python
student_data = pd.read_csv(student_data_to_load)
student_data.head()
```

```python
# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

Local Gorvernment Area Summary

```python
#Total number of unique schools
Unique_schools=school_data_complete["school_name"].nunique()
print(f"Total number of unique schools: {Unique_schools}")
```

```python
#Total number of students
Total_students_number = len(student_data)
print(f"Total number of students: {Total_students_number}")
```

```python
#Total budget
Total_budget = school_data["budget"].sum()
print(f"Total budget: {Total_budget}")
```

```python
#Average maths score
Average_maths_score = student_data["maths_score"].mean()
print(f"The average maths score: {Average_maths_score}")
```

```python
#Average reading score
Average_reading_score = student_data["reading_score"].mean()
print(f"The average reading score: {Average_reading_score}")
```

```python
#% passing maths (the percentage of students who passed maths)
Num_passing_maths = school_data_complete[(school_data_complete["maths_score"]>=50)].count()["student_name"]
Percentage_passing_maths = Num_passing_maths/(Total_students_number)*100
print(f"The percentage of students who passed maths: {Percentage_passing_maths}")
```

```python
#% passing reading (the percentage of students who passed reading)
Num_passing_reading = school_data_complete[(school_data_complete["reading_score"]>=50)].count()["student_name"]
Percentage_passing_reading = Num_passing_reading/(Total_students_number)*100
print(f"The percentage of students who passed reading: {Percentage_passing_reading}")
```

```python
#% overall passing (the percentage of students who passed maths AND reading
Num_passing_maths_reading = school_data_complete[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)].count()["student_name"]
#print(Num_passing_maths_reading)
Percentage_passing_maths_reading = Num_passing_maths_reading/(Total_students_number)*100
print(Percentage_passing_maths_reading)
```

```python
area_summary = pd.DataFrame(
    {"Total Schools": [Unique_schools],
     "Total Students": [Total_students_number],
     "Total Budget": [Total_budget],
     "Average Maths Score": [Average_maths_score],
     "Average Reading Score": [Average_reading_score],
     "% Passing Maths": [Percentage_passing_maths],
     "% Passing Reading":[Percentage_passing_reading],
     "% Overall Passing":[Percentage_passing_maths_reading]
     }
)

area_summary["Total Students"] = area_summary["Total Students"].map("{:,}".format)
area_summary["Total Budget"] = area_summary["Total Budget"].map("${:,.2f}".format)

area_summary
```

# School Summary

```python
# Using code provided to select the type per school from school_data
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
#Calculate the total student count per school from school_data
per_school_counts = school_data_complete.groupby("school_name").size()
print(per_school_counts)
```

```python
# Calculate the total school budget and per capita spending per school from school_data
#per_school_budget = school_data [
per_school_budget = school_data[["school_name","budget", "size"]].sort_values(by=['school_name'])
#print(per_school_budget)

#per_school_capita
#Make a new dataframe while keeping titles, then add column
per_school_capita = per_school_budget.loc[:, ["school_name","budget","size"]].sort_values(by=['school_name'])
per_school_capita["Per Student Budget"] = per_school_budget["budget"]/ per_school_budget["size"]
per_school_capita
```

```python
# Calculate the average test scores per school from school_data_complete
per_school_maths = school_data_complete.groupby(["school_name"])[["maths_score"]].mean()
per_school_maths
```

```python
# Calculate the average test scores per school from school_data_complete
#per_school_reading["Average Reading Score"] = school_data_complete.groupby(["school_name"])[["reading_score"]].mean()
per_school_reading = school_data_complete.groupby(["school_name"])[["reading_score"]].mean()
per_school_reading
```

```python
# Get the students who passed maths and passed reading by creating separate filtered DataFrames from school_data_complete.
school_passing_maths = school_data_complete.loc[(school_data_complete["maths_score"]>=50), :]
school_passing_maths
```

```python
school_passing_reading = school_data_complete.loc[(school_data_complete["reading_score"]>=50), :]
school_passing_reading
```

```python
# Get the students who passed both reading and maths in a separate DataFrame from school_data_complete.
passing_maths_and_reading = school_data_complete.loc[(school_data_complete["reading_score"]>=50) & (school_data_complete["maths_score"]>=50), :]
passing_maths_and_reading
```

```python
#  Calculate the Percentage Pass Rates
per_school_passing_maths = school_passing_maths.groupby(["school_name"]).agg(student_count_passing=('student_name', 'count'), total_students=('size', 'first'))
per_school_passing_maths["% Passing Maths"] = (per_school_passing_maths["student_count_passing"] / per_school_passing_maths["total_students"])*100
per_school_passing_maths
```

```python
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).agg(student_count_passing=('student_name', 'count'), total_students=('size', 'first'))
per_school_passing_reading["% Passing Reading"] = (per_school_passing_reading["student_count_passing"] / per_school_passing_reading["total_students"])*100
per_school_passing_reading
```

```python
overall_passing_rate = passing_maths_and_reading.groupby(["school_name"]).agg(school_type=('type', 'first'),student_count_passing=('student_name', 'count'), total_students=('size', 'first'))
overall_passing_rate["% Overall Passing"] = (overall_passing_rate["student_count_passing"] / overall_passing_rate["total_students"])*100
overall_passing_rate
```

```python
per_school_summary_merge = pd.merge(per_school_capita, per_school_maths, on="school_name")
per_school_summary_merge = pd.merge(per_school_summary_merge, per_school_reading, on="school_name")
per_school_summary_merge = pd.merge(per_school_summary_merge, per_school_passing_maths, on="school_name")
per_school_summary_merge = pd.merge(per_school_summary_merge, per_school_passing_reading, on="school_name")
per_school_summary_merge = pd.merge(per_school_summary_merge, overall_passing_rate, on="school_name")

per_school_summary_merge
```

```python
#specifying the columns to select
columns_to_select = ["school_name", "school_type", "total_students", "budget", "Per Student Budget", "maths_score", "reading_score", "% Passing Maths", "% Passing Reading", "% Overall Passing"]
#new dataframe
per_school_summary = per_school_summary_merge.loc[:, columns_to_select]
#rename columns
per_school_summary.rename(columns={"school_type": "School Type", "total_students":"Total Students", "budget":"Total School Budget", "maths_score":"Average Maths Score","reading_score":"Average Reading Score"}, inplace=True)
# Reset the index and drop the index column
per_school_summary.set_index('school_name', inplace=True) 
# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary
```

## Top Performing Schools (By % Overall Passing) 

```python
# Sort and show top five schools
top_schools = per_school_summary.sort_values(by='% Overall Passing', ascending=False)
top_schools.head(5)
```

```python
# Sort and show bottom five schools
bottom_schools = per_school_summary.sort_values(by='% Overall Passing', ascending=True)
bottom_schools.head(5)
```

## Maths Scores by Year

```python
# Create data series of scores by year levels using conditionals
year_nine = school_data_complete[(school_data_complete["year"] == 9)]
year_ten = school_data_complete[(school_data_complete["year"] == 10)]
year_eleven = school_data_complete[(school_data_complete["year"] == 11)]
year_twelve = school_data_complete[(school_data_complete["year"] == 12)]

# Group each by school name
year_nine_scores =year_nine.groupby('school_name')["maths_score"].agg([('Year 9','mean')])
year_ten_scores =year_ten.groupby('school_name')["maths_score"].agg([('Year 10' ,'mean')])
year_eleven_scores =year_eleven.groupby('school_name')["maths_score"].agg([('Year 11' ,'mean')])
year_twelve_scores =year_twelve.groupby('school_name')["maths_score"].agg([('Year 12' ,'mean')])

# Combine series into single DataFrame
maths_scores_by_year = pd.concat([year_nine_scores, year_ten_scores, year_eleven_scores, year_twelve_scores], axis=1)

# Minor data wrangling
maths_scores_by_year.index.name = None

# Display the DataFrame
maths_scores_by_year
```

## Reading Score by Year

```python
# Create data series of scores by year levels using conditionals
year_nine = school_data_complete[(school_data_complete["year"] == 9)]
year_ten = school_data_complete[(school_data_complete["year"] == 10)]
year_eleven = school_data_complete[(school_data_complete["year"] == 11)]
year_twelve = school_data_complete[(school_data_complete["year"] == 12)]

# Group each by school name
year_nine_scores =year_nine.groupby('school_name')["reading_score"].agg([('Year 9','mean')])
year_ten_scores =year_ten.groupby('school_name')["reading_score"].agg([('Year 10' ,'mean')])
year_eleven_scores =year_eleven.groupby('school_name')["reading_score"].agg([('Year 11' ,'mean')])
year_twelve_scores =year_twelve.groupby('school_name')["reading_score"].agg([('Year 12' ,'mean')])

# Combine series into single DataFrame
reading_scores_by_year = pd.concat([year_nine_scores, year_ten_scores, year_eleven_scores, year_twelve_scores], axis=1)

# Minor data wrangling
reading_scores_by_year.index.name = None

# Display the DataFrame
reading_scores_by_year
```

## Scores by School Spending

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
#  This step can be skipped but its best to make a copy.
school_spending_df = per_school_summary
school_spending_df
```

```python
# Categorise spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"].replace('[\\$]', '', regex=True).astype(float), spending_bins, labels=group_names, include_lowest=True)
school_spending_df
```

```python
#  Calculate averages for the desired columns.
spending_maths_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_maths = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Maths"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary =pd.concat([spending_maths_scores, spending_reading_scores, spending_passing_maths, spending_passing_reading, overall_passing_spending], axis=1)
# Display results
spending_summary
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], size_bins, labels=group_names, include_lowest=True)
per_school_summary
```

```python
# Calculate averages for the desired columns.
size_maths_scores = per_school_summary.groupby(["School Size"])["Average Maths Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_maths = per_school_summary.groupby(["School Size"])["% Passing Maths"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
size_summary =pd.concat([size_maths_scores, size_reading_scores, size_passing_maths, size_passing_reading, size_overall_passing], axis=1)
# Display results
size_summary
```
